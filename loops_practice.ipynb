16.Reverse the given number
input:234
output:432

```python
n=int(input())
rev=0
while n!=0:
  rem=n%10
  rev=rev*10+rem
  n=n//10
print(rev)
```

17.Sum of odd digits and product of even digits and their absolute difference
input:3456
output:16

```python
n=int(input())
even=1
odd=0
while n!=0:
  rem=n%10
  if rem%2==0:
    even=even*rem
  else:
    odd=odd+rem
  n=n//10
print(abs(even-odd))
```

18.Find factorial of N using a loop.Handle N=0.

21.Check if a number is a prime or not

```python
n=int(input())
ans=True
for i in range(2,n):
  if n%i==0:
    ans=False
    break
if ans:
  print("Prime")
else:
  print("Not a Prime")
```
