Simple Calculator using switch-case:read two numbers and an operator.Handle division by zero.

```python
a,b = map(int,input().split())
opr = input()
if opr == '+':
  print(a + b)
elif opr == '-':
  print(a - b)
elif opr == '*':
  print(a * b)
elif opr == '/':
  if b != 0:
    print(a / b)
  else:
    print("Error: Division by zero")
elif opr == '%':
  if b != 0:
    print(a % b)
  else:
    print("Error: Modulo by zero")
else:
    print("Invalid operator")
```

Type of triangle

```python
a,b,c=map(int,input().split())
if a==b==c:
  print("equilateral triangle")
elif a==b or b==c or a==c:
  print("isosceles triangle")
else:
  print("scalene triangle")
```
