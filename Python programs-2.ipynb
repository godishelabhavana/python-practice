Income tax calculator with slabs:0-2.5L=0%,2.5-5L=5%,5-10L=20%,10L=30%,print income tax and net income.

```python
income=int(input())
tax=0
net_income=income
if income<=250000:
  tax=0
elif income<=500000:
  tax=income-250000*0+(income-250000)*0.05
elif income<=1000000:
  tax=income-250000*0 + 500000*0.05 + (income-500000)*0.02
elif income>100000:
  tax=250000*0+500000*0.05+500000*0.02+(income-1000000)*0.3
net_income=income-tax
print("Income tax=",tax)
print("Net income=",net_income)
```

Determine the quadrant of a point(x,y) in a 2D plane.Handle axis cases(x=0 or y=0)

```python
x,y=map(int,input().split())
if x==0 and y==0:
  print("origin")
elif x==0:
  print("Y-axis")
elif y==0:
  print("X-axis")
elif x>0 and y>0:
  print("Q1")
elif x<0 and y>0:
  print("Q2")
elif x<0 and y<0:
  print("Q3")
elif x>0 and y<0:
  print("Q4")
else:
  print("Invalid")
```
