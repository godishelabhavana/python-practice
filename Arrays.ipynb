1.Read N integers into array.Print all elements.

2.Sum and Average of Array elements.

```python
n = int(input())
arr=list(map(int,input().split()))[:n]
sum=0
avg=0
for i in range (len(arr)):
  sum+=arr[i]
avg=sum/len(arr)
print(sum,avg)
```

3.Maximum and Minimum element in a Array.





```python
n=int(input())
arr=list(map(int,input().split()))[:n]
min=arr[0]
max=arr[0]
for i in range(len(arr)):
    if arr[i]>max:
       max=arr[i]
    if arr[i]<min:
       min=arr[i]
print("maximum: ",max)
print("minimum: ",min)
```

4.Count Occurances of a given element in an Array.

```python
n=int(input())
arr=list(map(int,input().split()))
b=int(input())
count=0
for i in range(n):
  if arr[i]==b:
    count+=1
print(count)
```

5.Linear search:return index of target,or -1 if not found.

```python
n=int(input())
arr=list(map(int,input().split()))
target=int(input())
for i in range(n):
  if arr[i]==target:
    print(f"Found at index: ",i)
    break;
else:
   print("-1")
```

6.Reverse an Array in-place(no extra array,swap from both ends).

```python
n=int(input())
arr=list(map(int,input().split()))
for i in range(n//2):
  arr[i],arr[n-i-1]=arr[n-i-1],arr[i]
print(arr)
```

**Using two pointer

```python
n=int(input())
arr=list(map(int,input().split()))
left=0
right=n-1
while left<right:
  arr[left],arr[right]=arr[right],arr[left]
  left+=1
  right-=1
print(arr)
```
